####2. Print all the teams which have started playing between 1930-1980. use “Since/lastApp” column (5 points)

```
# Extracting first four digits from string as year

laliga['Debut']=laliga.Debut.apply(lambda x: str(x)[:4])

# To perform mathematical operation on Debut column, converting it into numeric

laliga['Debut'] = pd.to_numeric(laliga['Debut'])

# Printing team debuts in between 1930 - 1980
print (laliga[['Team','Debut']][laliga.Debut>=1930][laliga.Debut<=1980])
```

####3. Print the list of teams which came Top 5 in terms of points (2.5 points)

```
# Checking existing data types
laliga.dtypes
```

```
# Points column type is detected as object, converting it to numeric type

laliga[['Points', 'GoalsFor', 'GoalsAgainst', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 'GamesLost']] = laliga[['Points', 'GoalsFor', 'GoalsAgainst', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 'GamesLost']].apply(pd.to_numeric)
laliga.dtypes
```

####4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```
# defining a function:

def Goal_diff_count():
  laliga['Goal Difference'] = laliga['GoalsFor'] - laliga['GoalsAgainst']
  
  print(laliga.loc[:,['Team', 'Goal Difference']])
  
  print("Team with Maximum Number of Goal Difference:\n",
        laliga.loc[:,['Team', 'Goal Difference']].sort_values(by = ['Goal Difference'], ascending=False).head(1), "\n\n")
  
  print("Team with Minimum Number of Goal Difference:\n",
        laliga.loc[:,['Team', 'Goal Difference']].sort_values(by = ['Goal Difference'], ascending=True).head(1), "\n")
```

```
# calling a function
Goal_diff_count()
```

####5. Create a new column with the name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

```
# Calculating Winning Percentage
laliga['Winning Percentage'] = (laliga['GamesWon'] / laliga['GamesPlayed']) * 100

# Replacing potential NaN values with 0's
laliga.fillna(0, inplace=True)

# Print
laliga.loc[:,['Team', 'GamesWon', 'GamesPlayed', 'Winning Percentage']]
```

####6. Print the top 5 teams which have the highest Winning percentage

```
laliga.loc[:,['Team', 'Points', 'BestPosition', 'Winning Percentage']].sort_values(by = ['Winning Percentage'], ascending=False).head(5)
```
